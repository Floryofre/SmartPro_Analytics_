# Tablas de la Base de datos 

```python
# El nombre de la base de datos a la que le sumaremos nuestras tablas 
database = 'nba_henry'

# Creamos 10 tablas:

# string conexion
conn_str = (


    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=' + server + ';'  
    'DATABASE=' + database + ';' 
    'Trusted_Connection=yes;'

)

# Create a connection
conn = pyodbc.connect(conn_str, autocommit=True)

# Crear un cursor
cursor = conn.cursor()




cursor.execute(
'''
               
    CREATE TABLE team (
    id INT PRIMARY KEY,
    full_name VARCHAR(255),
    abbreviation VARCHAR(50),
    nickname VARCHAR(255),
    city VARCHAR(255),
    state VARCHAR(255),
    year_founded INT
)
''')


cursor.execute(
'''
CREATE TABLE teams_details (
    team_id INT PRIMARY KEY,
    abbreviation VARCHAR(10),
    nickname VARCHAR(50),
    yearfounded INT,
    city VARCHAR(100),
    arena VARCHAR(100),
    arenacapacity INT,
    owner VARCHAR(100),
    generalmanager VARCHAR(100),
    headcoach VARCHAR(100),
    dleagueaffiliation VARCHAR(100)
)
''')



cursor.execute(
'''
CREATE TABLE game(
    season_id BIGINT,  -- Identificador único para la temporada
    team_id_home BIGINT,  -- Identificador único del equipo local
    team_abbreviation_home VARCHAR(10),  -- Abreviatura del equipo local
    team_name_home VARCHAR(100),  -- Nombre completo del equipo local
    game_id BIGINT PRIMARY KEY,  -- Identificador único para el partido
    game_date DATE,  -- Fecha y hora del partido
    matchup_home VARCHAR(100),  -- Descripción del enfrentamiento del equipo local
    wl_home VARCHAR(2),  -- Resultado del partido para el equipo local
    min BIGINT,  -- Minutos jugados en el partido
    fgm_home FLOAT,  -- Tiros de campo realizados por el equipo local
    fga_home FLOAT,  -- Tiros de campo intentados por el equipo local
    fg_pct_home FLOAT,  -- Porcentaje de tiros de campo del equipo local
    fg3m_home FLOAT,  -- Triples realizados por el equipo local
    fg3a_home FLOAT,  -- Triples intentados por el equipo local
    fg3_pct_home FLOAT,  -- Porcentaje de triples del equipo local
    ftm_home FLOAT,  -- Tiros libres realizados por el equipo local
    fta_home FLOAT,  -- Tiros libres intentados por el equipo local
    ft_pct_home FLOAT,  -- Porcentaje de tiros libres del equipo local
    oreb_home FLOAT,  -- Rebotes ofensivos del equipo local
    dreb_home FLOAT,  -- Rebotes defensivos del equipo local
    reb_home FLOAT,  -- Rebotes totales del equipo local
    ast_home FLOAT,  -- Asistencias del equipo local
    stl_home FLOAT,  -- Robos del equipo local
    blk_home FLOAT,  -- Bloqueos del equipo local
    tov_home FLOAT,  -- Pérdidas de balón del equipo local
    pf_home FLOAT,  -- Faltas personales del equipo local
    pts_home FLOAT,  -- Puntos anotados por el equipo local
    plus_minus_home BIGINT,  -- Estadística de plus-minus para el equipo local
    video_available_home BIGINT,  -- Indicador de disponibilidad de video para el equipo local
    team_id_away BIGINT,  -- Identificador único del equipo visitante
    team_abbreviation_away VARCHAR(10),  -- Abreviatura del equipo visitante
    team_name_away VARCHAR(100),  -- Nombre completo del equipo visitante
    matchup_away VARCHAR(100),  -- Descripción del enfrentamiento del equipo visitante
    wl_away VARCHAR(2),  -- Resultado del partido para el equipo visitante
    fgm_away FLOAT,  -- Tiros de campo realizados por el equipo visitante
    fga_away FLOAT,  -- Tiros de campo intentados por el equipo visitante
    fg_pct_away FLOAT,  -- Porcentaje de tiros de campo del equipo visitante
    fg3m_away FLOAT,  -- Triples realizados por el equipo visitante
    fg3a_away FLOAT,  -- Triples intentados por el equipo visitante
    fg3_pct_away FLOAT,  -- Porcentaje de triples del equipo visitante
    ftm_away FLOAT,  -- Tiros libres realizados por el equipo visitante
    fta_away FLOAT,  -- Tiros libres intentados por el equipo visitante
    ft_pct_away FLOAT,  -- Porcentaje de tiros libres del equipo visitante
    oreb_away FLOAT,  -- Rebotes ofensivos del equipo visitante
    dreb_away FLOAT,  -- Rebotes defensivos del equipo visitante
    reb_away FLOAT,  -- Rebotes totales del equipo visitante
    ast_away FLOAT,  -- Asistencias del equipo visitante
    stl_away FLOAT,  -- Robos del equipo visitante
    blk_away FLOAT,  -- Bloqueos del equipo visitante
    tov_away FLOAT,  -- Pérdidas de balón del equipo visitante
    pf_away FLOAT,  -- Faltas personales del equipo visitante
    pts_away FLOAT,  -- Puntos anotados por el equipo visitante
    plus_minus_away BIGINT,  -- Estadística de plus-minus para el equipo visitante
    season_type VARCHAR(20)  -- Tipo de temporada (por ejemplo, "Regular", "Playoffs")
)
''')




cursor.execute(
'''
CREATE TABLE game_info (
    game_id BIGINT PRIMARY KEY,  -- Identificador único para el partido
    game_date DATE,  -- Fecha y hora del partido
    attendance INT,  -- Asistencia al partido
)
''')


cursor.execute(
'''
CREATE TABLE other_stats (
    game_id BIGINT PRIMARY KEY,  -- Identificador único para el partido
    league_id BIGINT,  -- Identificador único para la liga
    team_id_home BIGINT,  -- Identificador único del equipo local
    team_abbreviation_home VARCHAR(10),  -- Abreviatura del equipo local
    team_city_home VARCHAR(100),  -- Ciudad del equipo local
    pts_paint_home BIGINT,  -- Puntos en la pintura anotados por el equipo local
    pts_2nd_chance_home BIGINT,  -- Puntos de segunda oportunidad anotados por el equipo local
    pts_fb_home BIGINT,  -- Puntos de contraataque anotados por el equipo local
    largest_lead_home BIGINT,  -- Mayor ventaja alcanzada por el equipo local
    lead_changes BIGINT,  -- Número de cambios de liderato en el partido
    times_tied BIGINT,  -- Número de veces que el marcador estuvo empatado durante el partido
    team_turnovers_home FLOAT,  -- Pérdidas de balón del equipo local
    total_turnovers_home FLOAT,  -- Total de pérdidas de balón del equipo local
    team_rebounds_home FLOAT,  -- Rebotes del equipo local
    pts_off_to_home FLOAT,  -- Puntos del equipo local tras pérdidas del oponente
    team_id_away BIGINT,  -- Identificador único del equipo visitante
    team_abbreviation_away VARCHAR(10),  -- Abreviatura del equipo visitante
    team_city_away VARCHAR(100),  -- Ciudad del equipo visitante
    pts_paint_away BIGINT,  -- Puntos en la pintura anotados por el equipo visitante
    pts_2nd_chance_away BIGINT,  -- Puntos de segunda oportunidad anotados por el equipo visitante
    pts_fb_away BIGINT,  -- Puntos de contraataque anotados por el equipo visitante
    largest_lead_away BIGINT,  -- Mayor ventaja alcanzada por el equipo visitante
    team_turnovers_away FLOAT,  -- Pérdidas de balón del equipo visitante
    total_turnovers_away FLOAT,  -- Total de pérdidas de balón del equipo visitante
    team_rebounds_away FLOAT,  -- Rebotes del equipo visitante
    pts_off_to_away FLOAT,  -- Puntos del equipo visitante tras pérdidas del oponente
)
''')





cursor.execute(
'''
CREATE TABLE common_player_info (
    person_id BIGINT PRIMARY KEY,  -- Identificador único para la persona
    first_name VARCHAR(50),  -- Nombre de pila de la persona
    last_name VARCHAR(50),  -- Apellido de la persona
    display_first_last VARCHAR(100),  -- Nombre completo en formato "Nombre Apellido"
    display_last_comma_first VARCHAR(100),  -- Nombre completo en formato "Apellido, Nombre"
    display_fi_last VARCHAR(50),  -- Inicial del nombre seguido del apellido
    player_slug VARCHAR(100),  -- Slug del jugador (URL-friendly)
    birthdate DATE,  -- Fecha de nacimiento de la persona
    school VARCHAR(100),  -- Escuela del jugador
    country VARCHAR(50),  -- País de origen de la persona
    last_affiliation VARCHAR(100),  -- Última afiliación del jugador
    height VARCHAR(10),  -- Altura del jugador
    weight FLOAT,  -- Peso del jugador
    season_exp FLOAT,  -- Experiencia en temporadas del jugador
    jersey VARCHAR(10),  -- Número de camiseta del jugador
    position VARCHAR(50),  -- Posición del jugador
    rosterstatus VARCHAR(50),  -- Estado en la lista del jugador
    games_played_current_season_flag VARCHAR(10),  -- Indicador de juegos jugados en la temporada actual
    team_id BIGINT,  -- Identificador único del equipo
    team_name VARCHAR(100),  -- Nombre completo del equipo
    team_abbreviation VARCHAR(10),  -- Abreviatura del equipo
    team_code VARCHAR(100),  -- Código del equipo
    team_city VARCHAR(100),  -- Ciudad del equipo
    playercode VARCHAR(100),  -- Código del jugador
    from_year INT,  -- Año de inicio de la carrera del jugador
    to_year INT,  -- Año de finalización de la carrera del jugador
    dleague_flag VARCHAR(10),  -- Indicador de si el jugador estuvo en la liga de desarrollo
    nba_flag VARCHAR(10),  -- Indicador de si el jugador estuvo en la NBA
    games_played_flag VARCHAR(10),  -- Indicador de juegos jugados
    draft_year VARCHAR(10),  -- Año del draft del jugador
    draft_round VARCHAR(10),  -- Ronda del draft del jugador
    draft_number VARCHAR(10),  -- Número de selección del draft del jugador
    greatest_75_flag VARCHAR(10)  -- Indicador de si el jugador está en la lista de los 75 mejores
)
''')


cursor.execute(
'''
CREATE TABLE player (
    id BIGINT PRIMARY KEY,  -- Identificador único para el jugador
    full_name VARCHAR(100),  -- Nombre completo del jugador
    first_name VARCHAR(50),  -- Nombre de pila del jugador
    last_name VARCHAR(50),  -- Apellido del jugador
    is_active INT  -- Indicador de si el jugador está activo (1) o no (0)
)
''')



cursor.execute(
'''
CREATE TABLE inactive_player (
    player_id BIGINT,  -- Identificador único para el jugador
    game_id BIGINT,  -- Identificador único para el partido
    first_name VARCHAR(50),  -- Nombre de pila del jugador
    last_name VARCHAR(50),  -- Apellido del jugador
    jersey_num VARCHAR(30),  -- Número de camiseta del jugador
    team_id BIGINT,  -- Identificador único del equipo
    team_city VARCHAR(100),  -- Ciudad del equipo
    team_name VARCHAR(100),  -- Nombre completo del equipo
    team_abbreviation VARCHAR(10),  -- Abreviatura del equipo
   
);
''')


cursor.execute(
'''
CREATE TABLE draft_history (
    person_id BIGINT PRIMARY KEY,  -- Identificador único para la persona
    player_name VARCHAR(100),  -- Nombre completo del jugador
    season BIGINT,  -- Temporada en la que se realizó el draft
    round_number BIGINT,  -- Número de la ronda del draft
    round_pick BIGINT,  -- Número de selección en la ronda
    overall_pick BIGINT,  -- Número de selección general en el draft
    draft_type VARCHAR(50),  -- Tipo de draft
    team_id BIGINT,  -- Identificador único del equipo
    team_city VARCHAR(100),  -- Ciudad del equipo
    team_name VARCHAR(100),  -- Nombre completo del equipo
    team_abbreviation VARCHAR(10),  -- Abreviatura del equipo
    organization VARCHAR(100),  -- Organización del jugador antes del draft
    organization_type VARCHAR(50),  -- Tipo de organización (por ejemplo, "universidad", "equipo internacional")
    player_profile_flag INT  -- Indicador de perfil del jugador
)
''')



cursor.execute(
'''
CREATE TABLE draft_combine_stats (
    player_id BIGINT PRIMARY KEY,
    season INT,
    first_name VARCHAR(255),
    last_name VARCHAR(255),
    player_name VARCHAR(255),
    position VARCHAR(50),
    height_wo_shoes FLOAT,
    height_wo_shoes_ft_in VARCHAR(50),
    height_w_shoes FLOAT,
    height_w_shoes_ft_in VARCHAR(50),
    weight FLOAT,
    wingspan FLOAT,
    wingspan_ft_in VARCHAR(50),
    standing_reach FLOAT,
    standing_reach_ft_in VARCHAR(50),
    body_fat_pct FLOAT,
    hand_length FLOAT,
    hand_width FLOAT,
    standing_vertical_leap FLOAT,
    max_vertical_leap FLOAT,
    lane_agility_time FLOAT,
    modified_lane_agility_time FLOAT,
    three_quarter_sprint FLOAT,
    bench_press FLOAT,
    spot_fifteen_corner_left VARCHAR(50),
    spot_fifteen_break_left VARCHAR(50),
    spot_fifteen_top_key VARCHAR(50),
    spot_fifteen_break_right VARCHAR(50),
    spot_fifteen_corner_right VARCHAR(50),
    spot_college_corner_left VARCHAR(50),
    spot_college_break_left VARCHAR(50),
    spot_college_top_key VARCHAR(50),
    spot_college_break_right VARCHAR(50),
    spot_college_corner_right VARCHAR(50),
    spot_nba_corner_left VARCHAR(50),
    spot_nba_break_left VARCHAR(50),
    spot_nba_top_key VARCHAR(50),
    spot_nba_break_right VARCHAR(50),
    spot_nba_corner_right VARCHAR(50),
    off_drib_fifteen_break_left VARCHAR(50),
    off_drib_fifteen_top_key VARCHAR(50),
    off_drib_fifteen_break_right VARCHAR(50),
    off_drib_college_break_left VARCHAR(50),
    off_drib_college_top_key VARCHAR(50),
    off_drib_college_break_right VARCHAR(50),
    on_move_fifteen VARCHAR(50),
    on_move_college VARCHAR(50)
)
''')



# Cerrar la conexión
cursor.close()
conn.close()
```
